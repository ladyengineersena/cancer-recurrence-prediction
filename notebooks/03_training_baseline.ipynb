# 03 - Training Baseline

Train Cox and RSF baselines and compare validation c-index.

```python
from pathlib import Path
import numpy as np
from src.data_pipeline import DataConfig, load_and_prepare
from src.models import train_cox, train_rsf, predict_risk_cox, predict_risk_rsf
from lifelines.utils import concordance_index

cfg = DataConfig(data_dir="../data/sample")
ds = load_and_prepare(cfg)

cox = train_cox(ds.x_train, ds.durations_train, ds.events_train)
rsf = train_rsf(ds.x_train, ds.durations_train, ds.events_train)

risk_val_cox = predict_risk_cox(cox, ds.x_val)
risk_val_rsf = predict_risk_rsf(rsf, ds.x_val)

c_cox = concordance_index(ds.durations_val, -risk_val_cox, ds.events_val)
c_rsf = concordance_index(ds.durations_val, -risk_val_rsf, ds.events_val)

c_cox, c_rsf
```
