# 04 - Survival Model (DeepSurv)

Train a DeepSurv model and compare against Cox/RSF.

```python
from src.data_pipeline import DataConfig, load_and_prepare
from src.models import train_deepsurv, predict_risk_deepsurv
from lifelines.utils import concordance_index

cfg = DataConfig(data_dir="../data/sample")
ds = load_and_prepare(cfg)

model = train_deepsurv(ds.x_train, ds.durations_train, ds.events_train, max_epochs=10)
risk_val = predict_risk_deepsurv(model, ds.x_val)
concordance_index(ds.durations_val, -risk_val, ds.events_val)
```
